```python
import pandas as pd

targeted_data1 = pd.read_csv("targeted_dataset.csv")
targeted_data2 = pd.read_csv("targeted_dataset_2.csv")
```

```python
targeted_data1 = targeted_data1[["company", "description", "main_business_category", "predicted"]]
```

```python
targeted_data2 = targeted_data2[["company", "description", "main_business_category", "predicted"]]
targeted_data2
```

```python
targeted_data = pd.concat([targeted_data1, targeted_data2], axis = 0, ignore_index=True)
targeted_data
```

```python
df = pd.read_csv("tournament_hints_data.csv")
df
```

```python
df.columns
```

```python
targeted_data = targeted_data.rename(columns={'company': 'commercial_name'})
targeted_data
```

```python
# Performing an inner join on 'company_normalized' and 'description_normalized'
merged_df = pd.merge(df, targeted_data, how='inner', on=['commercial_name', 'description', 'main_business_category'])

# Select the columns you want to keep
merged_df = merged_df[['commercial_name', 'business_tags', 'short_description', 'description', 'main_business_category', 'predicted']]
merged_df
```

```python
round5 = merged_df
round5['concatenated_text'] = merged_df.apply(lambda x: ';'.join(x[col] for col in ['commercial_name', 'business_tags', 'short_description', 'description', 'main_business_category'] if pd.notna(x[col])), axis=1)
round5 = round5[['concatenated_text', 'predicted']]
round5
```

```python
round4 = merged_df
round4['concatenated_text'] = merged_df.apply(lambda x: ';'.join(x[col] for col in ['commercial_name', 'business_tags', 'short_description', 'description'] if pd.notna(x[col])), axis=1)
round4 = round4[['concatenated_text', 'predicted']]
round4
```

```python
round3 = merged_df
round3['concatenated_text'] = merged_df.apply(lambda x: ';'.join(x[col] for col in ['commercial_name', 'business_tags', 'short_description'] if pd.notna(x[col])), axis=1)
round3 = round3[['concatenated_text', 'predicted']]
round3
```

```python
round2 = merged_df
round2['concatenated_text'] = merged_df.apply(lambda x: ';'.join(x[col] for col in ['commercial_name', 'business_tags'] if pd.notna(x[col])), axis=1)
round2 = round2[['concatenated_text', 'predicted']]
round2
```

```python
round1 = merged_df
round1['concatenated_text'] = merged_df.apply(lambda x: ';'.join(x[col] for col in ['commercial_name'] if pd.notna(x[col])), axis=1)
round1 = round1[['concatenated_text', 'predicted']]
round1
```

```python
round5.to_csv("round5.csv", index=False)
round4.to_csv("round4.csv", index=False)
round3.to_csv("round3.csv", index=False)
round2.to_csv("round2.csv", index=False)
round1.to_csv("round1.csv", index=False)
```
